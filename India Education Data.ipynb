## PART 1: Introduction

Literacy and level of education are basic indicators of the level of development achieved by a society. Spread of literacy is generally associated with important traits of modern civilization such as modernization, urbanization, industrialization, communication and commerce. Literacy forms an important input in overall development of individuals enabling them to comprehend their social, political and cultural environment better and respond to it appropriately. Higher levels of education and literacy lead to a greater awareness and also contributes in improvement of economic and social conditions. It acts as a catalyst for social upliftment enhancing the returns on investment made in almost every aspect of development effort, be it population control, health, hygiene, environmental degradation control, employment of weaker sections of the society.
   
According to the Census 2001, as many as 560,687,797 persons in the country are literate. Of these 336,533,716 are males and 224, 154,081 are females. While the overall literacy rate works out to be 64.8 %, the male literacy rate is 75.3% and that for females is 53.7%, showing a gap of 21.6 percentage points between the sexes at the national level. The gap is more in the rural areas. In the urban areas, higher literacy rate has been recorded both for males and females and the difference among the sexes is lower (13 percentage points ). Kerala, Mizoram, Lakshadweep, Goa and Chandigarh occupy the top five positions in literacy while Dadra & Nagar Haveli, Uttar Pradesh, Jammu & Kashmir, Arunachal Pradesh, Jharkhand, and Bihar, are at bottom.
   
The literacy rates for rural population are the highest in Kerala, followed by Lakshadweep, Mizoram, Goa, and Delhi. Fourteen (14) States / Uts have recorded less than 60 percent rural Literacy rate.

```python
#Library required and Dataset is imported
#![image.png](attachment:image.png)
#We need matplot to plot the data in for Iris dataset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Let's import python first as it an important package for data processing
import pandas as pd

# This method produces a lot of warning, so let's import the package and ignore those warnings 
import warnings 
warnings.filterwarnings("ignore")

# This handles  the error of jyputer not plotting
%matplotlib inline
```

```python
# lets load the education data set using pandas
eduindia = pd.read_csv("C:\\Users\\pnkum\\Desktop\\Project\\IndiaEducationData\\final_data.csv") 

# Lets us see the first 10 entries of the file.
eduindia.head(10)
```

Before we continue to plot different graphs on this dataset, It's important to check if there are any NA or NaN. If we find any we need to remove.

```python
eduindia.isnull().values.any()
```

```python
india1 = eduindia.loc[eduindia['Type'].isin(['Rural'])]
```

First we will compare total female population with female with graduate degree. 

### What is a line chart?

A line chart or line graph is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.[1] It is a basic type of chart common in many fields. It is similar to a scatter plot except that the measurement points are ordered (typically by their x-axis value) and joined with straight line segments.

Source: https://en.wikipedia.org/wiki/Line_chart 

```python
fig,ax = plt.subplots(figsize=(10,10))
plt.plot(india1.Code, india1.TFemales, 'red')
plt.plot(india1.Code, india1.GFemales, 'blue')
plt.title('Total Females vs Female Graduate population')
plt.xlabel('State Code')
plt.ylabel('population')
plt.show()
```

We see that there is very low graduate female across the state in India, irrespective of the total female population.

To classify the result beter, we will classify region wise.

First let us check the states with region by plotting a bar graph and pie chart

### What is a barplot?
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

soruce : https://en.wikipedia.org/wiki/Bar_chart

```python
fig,ax = plt.subplots(figsize=(8,8))
india1['part'].value_counts(sort=False).plot(kind='bar',color = 'blue')
plt.title('Number of state per region',fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Number of State',fontsize=15)
```

### what is pie chart?

A pie chart (or a circle chart) is a circular statistical graphic which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area), is proportional to the quantity it represents. While it is named for its resemblance to a pie which has been sliced, there are variations on the way it can be presented.

Source : https://en.wikipedia.org/wiki/Pie_chart 

```python
fig,ax = plt.subplots(figsize=(8,8))
india1['part'].value_counts(sort=False).plot(kind='pie')
plt.title('Platform Distribution',fontsize=15)
plt.xlabel('Platform',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

```python
ipm2=india1.groupby(by='part').sum()[['GMales','GFemales']]
ipm2.plot(kind='bar',figsize=(12,8),grid=True,title='Matches wone by batting or fielding')
```

Lets try to plot a line graph comparing the graduate male vs graduate females.

This graph shows the gap between region and also between males and female population within region.

```python
ipm2=india1.groupby(by='part').sum()[['GMales','GFemales']]
ipm2.plot(kind='line',figsize=(12,8),grid=True,title='Graduate Male vs Graduate Female')
```

Using subplot we can compare all the values of female education like Total female, total literate female, literate female without education, female with Non-Technical diploma, Technical diploma and graduate female.

```python
fig,ax = plt.subplots(figsize=(10,10))
plt.subplot(3,2,1)
plt.plot(india1.Code, india1.TFemales, 'blue')
plt.title('Total Female')
plt.subplot(3,2,3)
plt.plot(india1.Code, india1.NTDFemales,'green')
plt.title('Total Female with Non Technical Diploma')
plt.subplot(3,2,2)
plt.plot(india1.Code, india1.LFemales, 'red')
plt.title('Total Literate Female')
plt.subplot(3,2,4)
plt.plot(india1.Code, india1.LWEFemales, 'brown')
plt.title('Total Female who are literate without education')
plt.subplot(3,2,5)
plt.plot(india1.Code, india1.TDFemales, 'black')
plt.title('Total Female with Technical Diploma')
plt.subplot(3,2,6)
plt.plot(india1.Code, india1.GFemales)
plt.title('Total Female With Graduation')

plt.tight_layout()

plt.show()
```

## Part 4: Seaborn

```python
sns.FacetGrid(india1, hue="AreaName", size=8) \
   .map(plt.scatter, "TFemales", "GFemales") \
   .add_legend()
```

```python
sns.jointplot(x="TFemales", y="GFemales", data=india1, size=8, kind='reg', color='blue')
```

```python
# we will use seaborn jointplot shows bivariate scatterplots and univariate histograms with Kernel density 
# estimation in the same figure
sns.jointplot(x="TFemales", y="GFemales", data=india1, size=6, kind='kde', color='blue', space=0)
```

```python
# We will use seaborn jointplot shows bivariate scatterplots and univariate histograms with hex graph in the same figure
sns.jointplot(x="TFemales", y="GFemales", data=india1, size=6, kind='hex', color='#800000')
```

```python
sns.boxplot(x="part", y="GFemales", data=india1)
```

```python
# We try to strip the plot and use jitter to see all the point for each species and save it in axes as ax
ax = sns.boxplot(x="part", y="GFemales", data=india1)
ax = sns.stripplot(x="part", y="GFemales", data=india1, jitter=True, edgecolor="gray")
```

```python
# We use swarmplot to plot the data
sns.swarmplot(x="part", y="GFemales", data=india1)
```

```python
# We use violinplot to plot the data
sns.violinplot(x="part", y="GFemales", data=india1, size=10)
```

```python
# Seaborn plot is useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(india1, hue="part", size=6) \
   .map(sns.kdeplot, "GFemales") \
   .add_legend()
```
